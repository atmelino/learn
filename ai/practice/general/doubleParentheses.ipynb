```python
# Source - https://stackoverflow.com/a/42874845
# Posted by Mahi, modified by community. See post 'Timeline' for change history
# Retrieved 2026-03-14, License - CC BY-SA 3.0

def func(a):
    print("outer function, parameter a=",a)
    def func2(b):
        print("inner function, parameter b=",b)
        return a + b
    return func2
```

```python
func3 = func(a=1)
print(func3)
func3(2)
```

```python
func(1)(2)
```

```python
intermediate = func(a=1)
x = intermediate(b=2)
print(x)
```
